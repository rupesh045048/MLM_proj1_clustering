```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Select numeric columns for analysis
numeric_cols = ['payment_installments', 'amount', 'freight_value', 'Age']
numeric_data = custom_df[numeric_cols]

# Outlier analysis
plt.figure(figsize=(16, 6))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, 4, i)
    sns.boxplot(y=custom_df[col])
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()

# Skewness and kurtosis analysis
for col in numeric_cols:
    skewness = skew(custom_df[col])
    kurt = kurtosis(custom_df[col])
    print(f"{col}: Skewness = {skewness}, Kurtosis = {kurt}")
```

```python
# Calculate the quartiles and IQR for each numeric column
Q1 = custom_df[numeric_cols].quantile(0.25)
Q3 = custom_df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Define a function to remove outliers
def remove_outliers(df):
    return df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Remove outliers from the DataFrame
custom_df_no_outliers = remove_outliers(custom_df[numeric_cols])

# Display the shape of the DataFrame before and after removing outliers
print("Shape before removing outliers:", custom_df.shape)
print("Shape after removing outliers:", custom_df_no_outliers.shape)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(custom_df_no_outliers[numeric_cols])

# Create a new DataFrame with the scaled data
scaled_df = pd.DataFrame(scaled_data, columns=numeric_cols)

# Display the first few rows of the scaled DataFrame
scaled_df
```

```python
wcss = []

# Define the range of clusters to try
n_clusters_range = range(1, 11)

# Fit KMeans and calculate WCSS for each value of K
for k in n_clusters_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(n_clusters_range, wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Curve for Optimal K')
plt.grid(True)
plt.show()
```

```python
  silhouette_scores = []
davies_bouldin_scores = []

# Define the range of clusters to try
n_clusters_range = range(2, 11)  # Start from 2 clusters

# Fit KMeans and calculate scores for each value of K
for k in n_clusters_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_df)

    # Calculate silhouette score
    silhouette_avg = silhouette_score(scaled_df, cluster_labels)
    silhouette_scores.append(silhouette_avg)

    # Calculate Davies-Bouldin score
    davies_bouldin_avg = davies_bouldin_score(scaled_df, cluster_labels)
    davies_bouldin_scores.append(davies_bouldin_avg)

# Plot the scores
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(n_clusters_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal K')

plt.subplot(1, 2, 2)
plt.plot(n_clusters_range, davies_bouldin_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Davies-Bouldin Score')
plt.title('Davies-Bouldin Score for Optimal K')

plt.tight_layout()
plt.show()

# Display the scores for all clusters
print("Silhouette Scores:")
for k, score in zip(n_clusters_range, silhouette_scores):
    print(f"Number of Clusters: {k}, Silhouette Score: {score}")

print("\nDavies-Bouldin Scores:")
for k, score in zip(n_clusters_range, davies_bouldin_scores):
    print(f"Number of Clusters: {k}, Davies-Bouldin Score: {score}")
```

```python
from sklearn.decomposition import PCA

# Apply PCA to reduce the dimensionality to 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_df)

# Fit KMeans with the optimal number of clusters
kmeans = KMeans(n_clusters=4, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_df)

# Create a DataFrame for the reduced data and cluster labels
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df['Cluster'] = cluster_labels

# Plotting the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df, palette='viridis', legend='full')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', c='red', s=100, label='Centroids')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clustering with 4 clusters (PCA-reduced data)')
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Apply PCA to reduce the dimensionality to 3D
pca_3d = PCA(n_components=3)
pca_result_3d = pca_3d.fit_transform(scaled_df)

# Fit KMeans with the optimal number of clusters
kmeans = KMeans(n_clusters=4, random_state=42)
cluster_labels_3d = kmeans.fit_predict(scaled_df)

# Create a DataFrame for the reduced data and cluster labels
pca_df_3d = pd.DataFrame(data=pca_result_3d, columns=['PC1', 'PC2', 'PC3'])
pca_df_3d['Cluster'] = cluster_labels_3d

# Plotting the clusters in 3D
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for each cluster
for cluster in range(4):
    cluster_data = pca_df_3d[pca_df_3d['Cluster'] == cluster]
    ax.scatter(cluster_data['PC1'], cluster_data['PC2'], cluster_data['PC3'], label=f'Cluster {cluster}')

# Plot centroids
centroids = kmeans.cluster_centers_
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], marker='x', c='black', s=100, label='Centroids')

# Customize plot
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('Clustering with 4 clusters (PCA-reduced data)')
ax.legend()
plt.show()
```

```python
# Reverse transform the centroids to the original scale
centroids_original = scaler.inverse_transform(kmeans.cluster_centers_)

# Reverse transform the PCA-reduced data to the original scale
pca_inverse = pca.inverse_transform(pca_result)
pca_df_original = pd.DataFrame(data=pca_inverse, columns=scaled_df.columns)

# Assign cluster labels to the original data
pca_df_original['Cluster'] = cluster_labels

# Create separate DataFrames for each cluster
cluster_dataframes = []
for cluster in range(4):
    cluster_data = pca_df_original[pca_df_original['Cluster'] == cluster].drop('Cluster', axis=1)
    cluster_data['Cluster'] = cluster  # Add Cluster column to each cluster DataFrame
    cluster_dataframes.append(cluster_data)

# Concatenate all cluster DataFrames into a single DataFrame
combined_df = pd.concat(cluster_dataframes)

# Print the first few rows of each cluster DataFrame
for i, cluster_df in enumerate(cluster_dataframes):
    print(f"Cluster {i} DataFrame:")
    print(cluster_df)
    print("\n")

```
