#Objective
To develop a customer segmentation to give recommendations like saving plans, loans, wealth management, etc. on target customer groups by doing clustering analysis by K-means and BIRCH .

#Data Description :
The sample Dataset summarizes the usage behavior of about 9000 active credit card holders during the last 6 months. The file is at a customer level with 18 behavioral variables.

1.	CUST_ID: Customer ID
2.	BALANCE: Balance amount
3.	BALANCE_FREQUENCY: How frequently the balance is updated, score between 0 and 1
4.	PURCHASES: Amount of purchases made
5.	ONEOFF_PURCHASES: Amount of one-off purchases
6.	INSTALLMENTS_PURCHASES: Amount of installment purchases
7.	CASH_ADVANCE: Cash in advance given by the user
8.	PURCHASES_FREQUENCY: How frequently purchases are being made, score between 0 and 1
9.	ONEOFF_PURCHASES_FREQUENCY: How frequently one-off purchases are being made, score between 0 and 1
10.	PURCHASES_INSTALLMENTS_FREQUENCY: How frequently purchases in installments are being made, score between 0 and 1
11.	CASH_ADVANCE_FREQUENCY: How frequently the cash in advance is being paid
12.	CASH_ADVANCE_TRX: Number of transactions made with "Cash in Advance"
13.	PURCHASES_TRX: Number of purchase transactions made
14.	CREDIT_LIMIT: Limit of credit card for user
15.	PAYMENTS: Amount of payments done by user
16.	MINIMUM_PAYMENTS: Minimum amount of payments made by user
17.	PRC_FULL_PAYMENT: Percent of full payment paid by user
18.	TENURE: Tenure of credit card service for user

#COMPARISON OF K-MEANS & BIRCH
#Memory Usage:

KMeans: The KMeans algorithm used approximately 419.39 MB of memory.

Birch: The Birch algorithm used approximately 756.38 MB of memory.
The Birch algorithm tends to use more memory compared to KMeans, which could be due to the nature of the algorithm and its hierarchical structure.

#Time Taken:
KMeans: The KMeans algorithm took approximately 1.42 seconds to complete.

Birch: The Birch algorithm took approximately 1.27 seconds to complete.
Both algorithms performed relatively similarly in terms of time taken, with Birch being slightly faster than KMeans in this particular run.

#Conclusion:
If memory usage is a concern and the dataset is large, KMeans may be a better choice as it consumes less memory.
If speed is more important and memory usage is not a constraint, Birch could be a suitable option as it can be faster in some cases.

Both algorithms have their strengths and weaknesses .

# Dimensionality reduction
Converting the DataFrame into 2D DataFrame for visualization

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_df
```

#**Hyperparameter tuning**

Finding 'k' value by Elbow Method

```python
inertia = []
range_val = range(1,15)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(scaled_df))
    inertia.append(kmean.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

```python

kmeans_model=KMeans(4)
kmeans_model.fit_predict(scaled_df)
pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

#Visualization

```python
plt.figure(figsize=(8,8))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df_kmeans,palette=['red','green','blue','black'])
plt.title("Clustering using K-Means Algorithm")
plt.savefig("Clustering using K-Means Algorithm.jpg")
plt.show()
```

```python
# find all cluster centers
cluster_centers = pd.DataFrame(data=kmeans_model.cluster_centers_,columns=[df.columns])
# inverse transform the data
cluster_centers = scalar.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data=cluster_centers,columns=[df.columns])
cluster_centers
```

```python
from sklearn.cluster import Birch

# Instantiate the BIRCH model
birch_model = Birch(n_clusters=4)

# Fit and predict the clusters
birch_labels = birch_model.fit_predict(scaled_df)

# Visualize the clusters
pca_df_birch = pd.concat([pca_df, pd.DataFrame({'cluster': birch_labels})], axis=1)
sns.scatterplot(x='PCA1', y='PCA2', hue='cluster', data=pca_df_birch)
plt.title('BIRCH Clustering')
plt.savefig("Birch.jpg")
plt.show()
```

```python
cluster_sizes_birch = pd.Series(birch_labels).value_counts()
print("Cluster Sizes for BIRCH:")
print(cluster_sizes_birch)
```

```python
# Add the cluster labels to the original DataFrame
df_clustered_birch = pd.concat([df, pd.DataFrame({'cluster': birch_labels})], axis=1)

# Group the DataFrame by cluster and calculate the mean for each feature
cluster_means = df_clustered_birch.groupby('cluster').mean()

# Display the mean values for each cluster
print("Cluster Means for BIRCH:")
cluster_means


```

```python
# Visualize the clusters using box plots for each feature
plt.figure(figsize=(20, 12))
for i, col in enumerate(df.columns):
    plt.subplot(4, 5, i+1)
    sns.boxplot(x='cluster', y=col, data=df_clustered_birch)
    plt.title(col)
plt.tight_layout()
plt.show()
```
